Interesting observations:
Whatever coding format the actors names are in can't handle special characters which throws off many of the names. We would have to decrpyt/translate these names at some point.

Robert De Niro appears in the most IMDB 1000 movies by far. The average for an actor on the list is 1.65 and he's in 18 and the lead in 13 of those (without being in any series).

```python
#split the list of actors into three columns

a = movies.apply(lambda x: pd.Series(x.actors_list.strip('[]').strip('u').split(", u")[0]),axis=1).stack().reset_index(level=1, drop=True)
a.name = 'actor_1'
b = movies.apply(lambda x: pd.Series(x.actors_list.strip('[]').strip('u').split(", u")[1]),axis=1).stack().reset_index(level=1, drop=True)
b.name = 'actor_2'
c = movies.apply(lambda x: pd.Series(x.actors_list.strip('[]').strip('u').split(", u")[2]),axis=1).stack().reset_index(level=1, drop=True)
c.name = 'actor_3'
separateactors = movies.drop('actors_list', axis=1).join([a,b,c])
separateactors.head()
```

```python
separateactors.actor_1.value_counts()
```

```python
# print(movies.actors_list[0].split("'")[1])
# print(movies.actors_list[0].split("'")[3])
# print(movies.actors_list[0].split("'")[5])
```

```python
# df = imdb dataframe

#create blank list
actors_split = []

# iterate through the index values
for i in movies.index.values:
    #iterate through actors in each row, with a little processing
    for actor in movies.actors_list[i].strip("[]").strip('u').split(", u"):
        # create a blank dict for each row
        row = {}
        # add values to dict
        row['key'] = i
        row['actor'] = actor
        #add the row to the master list
        actors_split.append(row)
        
# merge the master and the original df together, matching the original df's index to the 'key' in our master list
# then dropping the un-needed cols
new_movies = movies.merge(pd.DataFrame(actors_split),left_index=True,right_on=['key']).drop(['actors_list','key'],axis=1)
new_movies.head()
```

```python
total_rating_pts = new_movies.groupby('actor').star_rating.sum().sort_values(ascending=False)
total_movies = new_movies.actor.value_counts()
total_movies.head()
```
